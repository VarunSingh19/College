```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1st Practical
a = [5,2,9,4,7]
b = [10,5,8,4,2]
plt.plot(a,b)
plt.xlabel("X-axis")
plt.ylabel("Y-axix")
plt.title("My 1st Gragh")
plt.show()
```

```python
# 2nd Practical
a = [1,2,3,4,5,6,7,8]
b = [0,0.6,0.2,15,10,8,16,21]
plt.plot(a)
plt.plot(b,"or")
plt.plot(list(range(0,22,3)))
plt.xlabel('Day --->')
plt.ylabel('Temp --->')
c = [4,2,6,8,3,20,13,15]
plt.plot(c,label = '4th Rep')
plt.legend(['1st Rep', '2nd Rep','3rd Rep','4th Rep'])
plt.title('All Features Discussed')
plt.show()
```

```python
# 3rd Practical
a = [5,6,8,3]
mylabels = ['apples','bananas','cherries','dates']
myexplode = [0.2,0.1,0.2,0.1]
plt.pie(a,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
# Reading CSV
df  = pd.read_csv("https://raw.githubusercontent.com/lgreski/pokemonData/master/Pokemon.csv")
df
# print(df.columns)
```

```python
# Using plot on df (csv file)
plt.bar(df['Attack'], df['Defense'])
plt.show()
```

```python
plt.scatter(df['Attack'], df['Defense'])
plt.show()
```

```python
plt.fill_between(df['Sp. Atk'].head(), df['Speed'].head())
plt.show()
```

```python
# plt.boxplot(df['HP'].head())
# plt.show()

columns_to_plot = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Create a boxplot for the selected columns
plt.boxplot(df[columns_to_plot].head())
plt.xticks(range(1, len(columns_to_plot) + 1), columns_to_plot)  # Add x-axis labels
plt.title('Boxplot for Pokemon Stats')
plt.show()
```
